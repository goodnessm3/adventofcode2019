```python
func_dict = {}
args_dict = {}

def register(func):
    '''decorator - functions are looked up by string
    docstring of each operation is its opcode'''
    code = func.__doc__
    func_dict[code] = func
    if func.__defaults__:
        args_dict[code] = len(func.__defaults__)
    else:
        args_dict[code] = 0
    return func

def dnf(func):
    '''destination not final, always assume that the final parameter is
    a memory address unless the function has this attribute, then pay
    attention to the mode as normal if it does have this as True'''
    func.dnf = True
    return func
    
def output(*args):
    print("OUTPUT:", args)
    
@register
def add(t, a=None, b=None, dest=None):
    '''01'''
    t[dest] = a + b

@register
def mult(t, a=None, b=None, dest=None):
    '''02'''
    t[dest] = a * b
    
@register
def store(t, dest=None):
    '''03'''
    a = input(">")  # get input from user
    #print(f"stored {a} at position {dest}")
    t[dest] = int(a)
    
@register
def retrieve(t, dest=None):
    '''04'''
    output(t[dest])
    
@register
@dnf
def jump_if_true(t, a=None, b=None):
    '''05'''
    global current_pos
    global instruction_pointer_changed
    
    if not a == 0:
        current_pos = b
        instruction_pointer_changed = True
        
@register
@dnf
def jump_if_false(t, a=None, b=None):
    '''06'''
    global current_pos
    global instruction_pointer_changed
    
    if a == 0:
        current_pos = b
        instruction_pointer_changed = True
        
@register
def less_than(t, a=None, b=None, dest=None):
    '''07'''
    if a < b:
        t[dest] = 1
    else:
        t[dest] = 0
        
@register
def equals(t, a=None, b=None, dest=None):
    '''08'''
    if a == b:
        t[dest] = 1
    else:
        t[dest] = 0
    
@register
def halt(t):
    '''99'''
    return True  # the only function that returns
    


```

```python
def prepare_args(t, modes, args, func):
    
    '''reads the required number of args, looks them up if necessary,
    returns a list to be passed to the instruction to be run'''
    # modes is a string of "10010"...
    
    out = []
    #print(f"preparing args, modes:{modes}, args:{args}")
    
    xx = 0  # loop counter
    
    for m, a in zip(modes, args): 
        if xx+1 == len(args):
            try:
                qz = func.dnf  # (memory) destination not final (argument), treat final arg as normal
            except AttributeError:         
                out.append(a) # final argument is always an address to put something
                break
        if m == "0":  # positional mode
            out.append(t[a])  # lookup value by location
        elif m == "1":  # immediate mode
            out.append(a)  # take the argument itself as the value
        xx += 1
            
    #print(f"returning args:{out}")
    return out
```
